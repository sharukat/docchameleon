```python
import time
import pandas as pd
from tqdm import tqdm
import lib.global_settings as settings
from ShaDoC.search import course_urls_retriever

import os
from dotenv import load_dotenv
load_dotenv(dotenv_path="/Users/sharukat/Documents/ResearchYU/Code/doc-customizer-llm/doc_customizer_llm/.env")
os.environ["COHERE_API_KEY"] = os.getenv("COHERE_API_KEY")
```

```python
df = pd.read_csv(os.path.join(settings.DATA_PATH, "question_retrieval_eval_dataset.csv"))
df = df[["QuestionId", "QuestionAPI", "IssueType", "Title", "Question", "GroundTruth"]]
df = df[28:]
df
```

```python
results = []
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    courses, yt = course_urls_retriever(row['Question'])
    time.sleep(8)
    res = {"course_urls": courses, "yt_urls": yt}
    results.append(res)
```
