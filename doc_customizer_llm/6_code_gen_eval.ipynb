```python
import os
import pandas as pd
from dotenv import load_dotenv
load_dotenv(dotenv_path="/Users/sharukat/Documents/ResearchYU/Code/doc-customizer-llm/doc_customizer_llm/.env")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

from human_eval.data import read_problems
import lib.global_settings as settings
```

```python
problems = read_problems()
problems = {key: problems[key] for key in list(problems.keys())[16:]}

eval_problems = []
for key, value in problems.items():
    # print(value)
    eval_problems.append({
        "task_id": key, 
        "prompt": value["prompt"], 
        "canonical_solution": value["canonical_solution"],
        "test": value["test"],
        })

df = pd.DataFrame(eval_problems)
df
```

```python
import requests
import pandas as pd
import json
import ast


endpoint = "https://sharukat--code-langchain-serve-dev.modal.run/codelangchain/invoke"
index = 0
records = []
for key, value in problems.items():
    input = {
        "problem": value,
        "prompt": value["prompt"],
        # "question_id": df["QuestionId"][index],
        # "title": df["Title"][index], 
        # "api_name": df["QuestionAPI"][index],
        # "issue_type": df["ResponseType"][index],
        # "question": df["Question"][index],   
        # "ground_truth": df["GroundTruth"][index]
        }

    data = {
        "input": str(input),
        "recursion_limit": 50,
        "kwargs": {},
    }
    try:
        response = requests.post(
            url=endpoint,
            json=data,
        )
        results = json.loads(response.text)
        results =ast.literal_eval(results['output'])
        input.update(results)
        records.append(input)    
    except:
        print(value["task_id"])
        pass 

    break
```

```python
res = pd.DataFrame(records)
# res = res.drop(index=[4, 6,7, 10, 11, 12])
res
```
