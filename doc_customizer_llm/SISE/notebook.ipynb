```python
from sklearn.utils import resample

# Separate majority and minority classes
train_df_majority = train_df[train_df['isMeaningful'] == 'no']
train_df_minority = train_df[train_df['isMeaningful'] == 'yes']

df_majority_downsampled = resample(train_df_majority,
                                   replace=False,    # sample without replacement
                                   n_samples=len(train_df_minority), # to match minority class
                                   random_state=42)

train_balanced = pd.concat([df_majority_downsampled, train_df_minority])
train_balanced = train_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
train_balanced.to_csv('/Users/sharukat/Documents/ResearchYU/Code/doc-customizer-llm/doc_customizer_llm/SISE/train_balanced.csv', index=False)
train_balanced
```
