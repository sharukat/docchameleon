````python
import random 

df = pd.read_csv(os.path.join(settings.DATA_PATH, "eval_data_ragas.csv"))
q_ids = df['QuestionId'].tolist()
q_ids = q_ids[:50]

hashmap = {}
for id in q_ids:
    hashmap[id] = 3


def assign_ids(user_hash, hashmap):
    if len(user_hash) == 5:
        return user_hash
    
    id = random.choice(list(hashmap.keys()))
    if hashmap[id] == 0:
        del hashmap[id]
        return assign_ids(user_hash, hashmap)
    else:
        if id not in user_hash:
            user_hash.add(id)
            hashmap[id] -= 1
        return assign_ids(user_hash, hashmap)


assignments = []
for i in range(30):
    user_hash = set()
    user_hash = assign_ids(user_hash, hashmap)
    assignments.append({'user':i, 'assigned_ids': list(user_hash)})

print(assignments)

````

````python


# LANGCHAIN MODULES
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field, validator
from langchain_core.output_parsers import JsonOutputParser

import os
from dotenv import load_dotenv
load_dotenv(dotenv_path="/Users/sharukat/Documents/ResearchYU/Code/doc-customizer-llm/doc_customizer_llm/.env")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

filename = "tf.gather"
filename_with_ext = os.path.join("/Users/sharukat/Documents/ResearchYU/Code/doc-customizer-llm/doc_customizer_llm/ShaDoC/docs_samp", filename) + ".md" 
with open(filename_with_ext, "r", encoding="utf-8") as file:
    content = file.read()

class Output(BaseModel):
    summary: str = Field(description="Summary of the API documentation")

def question_intent_identifier(documentation):
    print(f"Summarizing...")

    parser = JsonOutputParser(pydantic_object=Output)
    format_instructions = parser.get_format_instructions()

    template = """
        You are an expert in summarizing API documentation. 
        Your task is to isummarize the API documentation without chainging the code snippets, keywords or arguments.
        And do not change any content within markdown tags. Summarize only the text which are not enclosed with any tags.
        Those unchanged content should be in the final output following the same structure.

        Documentation:
        {documentation}

        The output should be in the same markdown format.
        {format_instructions}
    """

    PROMPT = PromptTemplate(
        template=template,
        input_variables=['documentation'], 
        partial_variables={"format_instructions":format_instructions})
    
    llm = ChatOpenAI(temperature=0, model_name="gpt-4o")
    chain = PROMPT | llm | parser

    prompt = {"documentation" : documentation}
    response = chain.invoke(prompt)
    print(f"Done")
    return response['summary']

output = question_intent_identifier(content)
````
